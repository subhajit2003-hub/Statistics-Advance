1. What is a random variable in probability theory?
 -> A random variable is a fundamental concept in probability theory, serving as a bridge between abstract probabilistic experiments and numerical outcomes.

2. What are the types of random variables?
-> In probability theory, random variables fall into three main types based on their range of values:

1. Discrete Random Variables
These take on a countable set of distinct values (finite or countably infinite). They’re described by a probability mass function (PMF).
Examples & common distributions:

* Bernoulli: two outcomes (0/1), e.g., coin flip
* Binomial: number of successes in n independent trials
* Geometric: trials until first success
* Poisson: count of rare events over an interval

2. Continuous Random Variables
These can assume any value within a real interval (uncountably infinite). Their probability is described by a probability density function (PDF)—the probability of any exact value is zero, but ranges have nonzero probability.
Examples & common distributions:

* Uniform: all values in [a, b] equally likely
* Normal (Gaussian): the familiar bell curve
* Exponential: time between Poisson events
* Others include log-normal, Weibull…

3. Mixed (Hybrid) Random Variables
These have both discrete and continuous components.
Example: A variable that is zero half the time (discrete point-mass at 0), but otherwise follows a continuous distribution.

3. What is the difference between discrete and continuous distributions?
-> | Feature                    | Discrete                          | Continuous                     |
| -------------------------- | --------------------------------- | ------------------------------ |
| Support                    | Countable values (e.g., integers) | Any real number in an interval |
| Probability of exact value |                                   | Always 0 for any single point  |
| Probability over range     | Sum of PMF values                 | Integral of PDF over interval  |
| Visualization              | Bars/Tables                       | Smooth density curve           |
| Common Examples            | Binomial, Poisson                 | Normal, Exponential, Uniform   |

21. What are the assumptions for applying the Central Limit Theorem?
-> The Central Limit Theorem (CLT) holds under a few key assumptions that ensure the sample mean (or sum) approximates a normal distribution:

1. Independence
The observations must be independent—one sample shouldn't influence another. For small populations and sampling without replacement, your sample should be no more than 10% of the population to approximate independence

2. Identical Distribution (i.i.d.)
Each sample should be drawn from the same population distribution. In the classical CLT, variables are independent and identically distributed with the same mean and variance .

More advanced versions (e.g., Lindeberg–Feller) allow non-identical distributions, provided no single variable dominates and certain technical conditions are met.

3. Finite Variance
Each variable must have a finite mean and variance—distributions with infinite variance (like the Cauchy) violate CLT.

4. Sufficiently Large Sample Size
The approximation improves with larger sample size, commonly n ≥ 30 is suggested as a rule-of-thumb. More skewed distributions may require even larger samples.

2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range [a, b]: integer values a, a+1, ..., b
a, b = 1, 6
dist = randint(low=a, high=b+1)  # SciPy's randint uses half-open [a, b+1)

# Values and their PMF
x = np.arange(a, b+1)
pmf = dist.pmf(x)

# Plot PMF using a bar chart
plt.figure(figsize=(6,4))
plt.bar(x, pmf, color='skyblue', edgecolor='k')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title(f'Discrete Uniform PMF on [{a}, {b}]')
plt.xticks(x)
plt.ylim(0, pmf.max() * 1.2)
plt.show()
```

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pmf(x, p):
    """
    x: observed value (0 or 1)
    p: probability of success (0 <= p <= 1)
    Returns the probability mass at x.
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0.0
for x in [0, 1, 2]:
    print(f"P(X={x}) =", bernoulli_pmf(x, 0.3))
```

4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n, p = 10, 0.5
num_samples = 1000  # simulate 1000 experiments

# Generate binomial samples: number of successes in each of the experiments
samples = np.random.binomial(n=n, p=p, size=num_samples)

# Plot the histogram of simulated data
plt.figure(figsize=(8, 4))
plt.hist(samples, bins=np.arange(n+2)-0.5, density=True,
         edgecolor='black', alpha=0.7)
plt.xlabel('Number of Successes')
plt.ylabel('Relative Frequency')
plt.title(f'Histogram of {num_samples} samples from Binomial(n={n}, p={p})')
plt.xticks(range(n+1))
plt.grid(axis='y', alpha=0.3)

# Optionally overlay the theoretical PMF
x = np.arange(n+1)
pmf = binom.pmf(x, n, p)
plt.scatter(x, pmf, color='red', zorder=5, label='Theoretical PMF')
plt.vlines(x, 0, pmf, colors='red', linestyles='dashed', zorder=4)
plt.legend()
plt.show()
```

5. Create a Poisson distribution and visualize it using Python.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# 1. Set the average rate (λ)
lam = 4  # mean number of events per interval

# 2. Generate possible event counts
k = np.arange(0, 15)

# 3. Compute theoretical PMF
pmf = poisson.pmf(k, mu=lam)

# 4. Simulate a sample
sample = poisson.rvs(mu=lam, size=10000)

# 5. Plot the PMF
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(k, pmf, 'bo-', label='Poisson PMF')
plt.xlabel('Number of events')
plt.ylabel('Probability')
plt.title(f'Theoretical PMF (λ = {lam})')
plt.grid(True)

# 6. Plot histogram of simulated data
plt.subplot(1, 2, 2)
counts, bins, patches = plt.hist(sample, bins=np.arange(k.min(), k.max() + 2) - 0.5,
                                 density=True, edgecolor='black', alpha=0.7)
plt.plot(k, pmf, 'ro', label='PMF overlay')
plt.xlabel('Number of events')
plt.ylabel('Relative frequency')
plt.title('Histogram of simulated Poisson data')
plt.legend()
plt.tight_layout()
plt.show()
```

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

def plot_discrete_uniform_cdf(a, b):
    """
    Plots the CDF for a discrete uniform distribution on integers {a, a+1, ..., b}.
    """
    # Create the distribution object
    dist = randint(a, b + 1)  # Note: high is exclusive

    # Generate x-values covering the full range
    x = np.arange(a - 1, b + 1)  # Extend one step before for plotting CDF=0 before start

    # Compute the CDF
    cdf_vals = dist.cdf(x)

    # Plot the CDF as a step function
    plt.figure(figsize=(6, 4))
    plt.step(x, cdf_vals, where='post', color='blue')
    plt.xlabel('x')
    plt.ylabel('F(x) = P(X ≤ x)')
    plt.title(f'Discrete Uniform CDF on [{a}, {b}]')
    plt.ylim(-0.05, 1.05)
    plt.grid(alpha=0.3)
    plt.xticks(np.arange(a - 1, b + 1))
    plt.show()

if __name__ == "__main__":
    plot_discrete_uniform_cdf(a=1, b=6)
```

7. Generate a continuous uniform distribution using NumPy and visualize it.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# 1. Define distribution limits
a, b = 0.0, 15.0  # range [a, b)

# 2. Generate random samples
n_samples = 4000
samples = np.random.uniform(low=a, high=b, size=n_samples)

# 3. Plot histogram (empirical density)
plt.figure(figsize=(4, 2))
count, bins, _ = plt.hist(samples, bins=50, density=True,
                          edgecolor='black', alpha=0.6,
                          label='Empirical density')

# 4. Overlay theoretical PDF
x = np.linspace(a, b, 300)
pdf = uniform.pdf((x - a) / (b - a)) / (b - a)  # or use uniform(loc=a, scale=b-a).pdf(x)
# Simplified: uniform.pdf(x, loc=a, scale=b-a)
pdf = uniform(loc=a, scale=b - a).pdf(x)

plt.plot(x, pdf, 'r--', linewidth=2, label='Theoretical PDF')

# 5. Final touches
plt.title(f'Continuous Uniform U({a}, {b})\nEmpirical vs Theoretical PDF')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

8. Simulate data from a normal distribution and plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Simulation parameters
mu, sigma = 0, 2        # mean and standard deviation
N = 80000               # number of samples

# 2. Generate random samples from N(mu, sigma^2)
data = np.random.normal(loc=mu, scale=sigma, size=N)  # uses numpy.random.normal :contentReference[oaicite:4]{index=4}

# 3. Plot the histogram (normalized to density)
count, bins, _ = plt.hist(data, bins=40, density=True, alpha=0.8, color='skyblue',
                          edgecolor='black', label='Simulated samples')

# 4. Prepare theoretical PDF values
x = np.linspace(bins.min(), bins.max(), 500)
pdf = norm.pdf(x, loc=mu, scale=sigma)

# 5. Overlay the theoretical PDF
plt.plot(x, pdf, 'r--', linewidth=2, label='Theoretical PDF')

# 6. Customize the plot
plt.title(f'Normal Distribution Samples (μ={mu}, σ={sigma})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def simulate_clt(pop_dist, sample_size, num_samples=10000):
    """
    Simulates CLT by drawing 'num_samples' of size 'sample_size' from pop_dist,
    computing sample means, and plotting their distribution vs Normal.
    """
    # Draw samples: result is num_samples × sample_size
    samples = pop_dist(size=(num_samples, sample_size))
    # Compute the mean of each sample
    sample_means = samples.mean(axis=1)

    # Plot histogram of sample means
    plt.figure(figsize=(8, 5))
    count, bins, _ = plt.hist(sample_means, bins=30, density=True,
                              alpha=0.6, color='skyblue', edgecolor='black',
                              label='Sample means')

    # Overlay Normal distribution
    mu, sigma = sample_means.mean(), sample_means.std()
    x = np.linspace(bins.min(), bins.max(), 200)
    plt.plot(x, norm.pdf(x, loc=mu, scale=sigma), 'r--',
             label=f'Normal fit: μ={mu:.2f}, σ={sigma:.2f}')

    plt.title(f'CLT: Uniform→Normal (n={sample_size})')
    plt.xlabel('Sample mean')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

# Example usage: Non-normal (Uniform) population
if __name__ == "__main__":
    # Uniform population on [0, 1)
    pop_dist = lambda size: np.random.uniform(0, 1, size)

    # Simulate for different sample sizes
    for n in [1, 5, 30, 100]:
        simulate_clt(pop_dist, sample_size=n)
```

11.Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, skew, kurtosis

# Parameters
population_mu, population_sigma = 5, 2  # normal population N(5, 2²)
sample_sizes = [1, 5, 30, 100]          # different sample sizes
num_samples = 10000                      # number of simulations per scenario

# Simulate and plot for each sample size
plt.figure(figsize=(12, 10))
for i, n in enumerate(sample_sizes, 1):
    # Draw samples: matrix of shape (num_samples, n)
    samples = np.random.normal(loc=population_mu,
                               scale=population_sigma,
                               size=(num_samples, n))
    means = samples.mean(axis=1)

    # Estimate mean and std of sample means
    est_mu, est_sigma = means.mean(), means.std(ddof=1)

    # Compute skewness & kurtosis
    s, k = skew(means), kurtosis(means, fisher=False)

    # Plot histogram and overlay theoretical normal
    plt.subplot(2, 2, i)
    count, bins, _ = plt.hist(means, bins=30, density=True,
                              alpha=0.6, color='skyblue',
                              edgecolor='black')
    x = np.linspace(bins.min(), bins.max(), 200)
    plt.plot(x, norm.pdf(x, loc=population_mu,
                         scale=population_sigma/np.sqrt(n)),
             'r--', linewidth=2, label='Theoretical')

    plt.title(f'n={n}: mean≈{est_mu:.2f}, σ≈{est_sigma:.2f}\n'
              f'skew={s:.2f}, kurtosis={k:.2f}')
    plt.xlabel('Sample Means')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

13. Generate random variables and calculate their corresponding probabilities using the binomial distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def simulate_and_plot_binomial(n=10, p=0.5, num_samples=1000):
    # 1. Simulate binomial samples (number of successes)
    samples = np.random.binomial(n, p, size=num_samples)

    # 2. Unique outcomes and their empirical frequencies
    values, counts = np.unique(samples, return_counts=True)
    freqs = counts / num_samples

    # 3. Calculate theoretical PMF for possible k = 0...n
    k = np.arange(n + 1)
    pmf_values = binom.pmf(k, n, p)

    # 4. Plot empirical histogram and theoretical PMF
    plt.figure(figsize=(8, 5))
    plt.bar(values - 0.2, freqs, width=0.4, color='skyblue', label='Empirical freq')
    plt.bar(k + 0.2, pmf_values, width=0.4, color='orange', alpha=0.7, label='Theoretical PMF')
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.xlabel('Number of Successes (k)')
    plt.ylabel('Probability / Frequency')
    plt.xticks(k)
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

    return samples, pmf_values
if __name__ == "__main__":
    samples, pmf = simulate_and_plot_binomial(n=10, p=0.35, num_samples=5000)
    print("Theoretical PMF:", np.round(pmf, 3))
```

15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from statsmodels.stats.weightstats import ztest

def one_sample_ztest(data, popmean, alpha=0.05):
    """
    Performs a one-sample z-test on data against a known population mean.
    Returns z_stat, p_value, and decision for a two-tailed test.
    """
    z_stat, p_value = ztest(data, value=popmean)
    print(f"Sample mean = {np.mean(data):.3f}, n = {len(data)}")
    print(f"Population mean (H0): {popmean}")
    print(f"Z-statistic = {z_stat:.3f}")
    print(f"P-value (two-tailed) = {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject H₀ at α = {alpha}. Evidence suggests mean ≠ {popmean}.")
    else:
        print(f"Fail to reject H₀ at α = {alpha}. No evidence mean differs from {popmean}.")

    return z_stat, p_value
if __name__ == "__main__":
    # Suppose we know the population mean μ=100 and population σ is known or sample size is large (Z-test appropriate) :contentReference[oaicite:0]{index=0}
    np.random.seed(42)
    data_sample = np.random.normal(102, 5, size=50)  # Simulated data
    z, p = one_sample_ztest(data_sample, popmean=100)
```

16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval of the mean for `data` using a t-distribution.
    Returns (lower_bound, upper_bound).
    """
    data = np.asarray(data)
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # standard error of the mean: s / sqrt(n)
    dof = n - 1
    interval = stats.t.interval(confidence, dof, loc=mean, scale=sem)
    return mean, interval

if __name__ == "__main__":
    # Sample data (e.g., weights or any metric)
    data = [12, 15, 14, 10, 13, 17, 14, 15, 16, 14]
    mean, (lower, upper) = confidence_interval(data, confidence=0.95)
    print(f"Sample mean = {mean:.3f}")
    print(f"95% Confidence interval = [{lower:.3f}, {upper:.3f}]")
```

17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def simulate_and_ci(mu=50, sigma=5, n=30, confidence=0.95):
    # 1. Generate normal data
    data = np.random.normal(loc=mu, scale=sigma, size=n)

    # 2. Calculate sample statistics
    mean = np.mean(data)
    sem = stats.sem(data)  # standard error: s / sqrt(n)
    df = n - 1

    # 3. Get t-critical value and compute CI
    t_crit = stats.t.ppf((1 + confidence) / 2, df=df)
    margin = t_crit * sem
    lower, upper = mean - margin, mean + margin

    print(f"Sample mean = {mean:.2f}")
    print(f"{int(confidence*100)}% CI = [{lower:.2f}, {upper:.2f}]")

    # 4. Plot histogram and CI lines
    plt.figure(figsize=(8, 4))
    plt.hist(data, bins=10, density=True, alpha=0.6, edgecolor='k', color='skyblue')
    plt.axvline(mean, color='black', label='Sample mean')
    plt.axvline(lower, color='red', linestyle='--', label='CI lower bound')
    plt.axvline(upper, color='red', linestyle='--', label='CI upper bound')
    plt.title(f'Normal Data (n={n}, μ={mu}, σ={sigma})')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

    return data, (lower, upper)

if __name__ == "__main__":
    np.random.seed(42)
    simulate_and_ci(mu=50, sigma=5, n=30, confidence=0.95)
```

19. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(mu=5, max_k=15):
    """
    Computes and plots the CDF of a Poisson(mu) from k=0 to max_k.
    """
    k = np.arange(0, max_k + 1)
    cdf_vals = poisson.cdf(k, mu)

    plt.figure(figsize=(8, 4))
    plt.step(k, cdf_vals, where='post', color='blue')
    plt.scatter(k, cdf_vals, color='red')
    plt.xlabel('k (number of events)')
    plt.ylabel('P(X ≤ k)')
    plt.title(f'Poisson CDF (λ = {mu})')
    plt.ylim(-0.02, 1.02)
    plt.grid(alpha=0.3)
    plt.show()

    return k, cdf_vals

if __name__ == "__main__":
    mu = 5
    k_values, cdf_values = plot_poisson_cdf(mu=mu, max_k=15)

    # Example interpretation:
    k0 = 7
    p_le_7 = cdf_values[np.where(k_values == k0)][0]
    print(f"P(X ≤ {k0}) ≈ {p_le_7:.3f} for Poisson(λ={mu})")
    print(f"→ There's about a {p_le_7*100:.1f}% chance of observing 7 or fewer events.")
```

21. Write a Python program to compare the standard deviations of two datasets and visualize the difference.

```python
import numpy as np
import matplotlib.pyplot as plt

def compare_std(data1, data2, labels=('Dataset 1', 'Dataset 2')):
    # 1. Convert to NumPy arrays
    data1 = np.asarray(data1)
    data2 = np.asarray(data2)

    # 2. Calculate mean and standard deviation
    mean1, std1 = np.mean(data1), np.std(data1, ddof=1)
    mean2, std2 = np.mean(data2), np.std(data2, ddof=1)

    print(f"{labels[0]} — mean: {mean1:.2f}, std: {std1:.2f}")
    print(f"{labels[1]} — mean: {mean2:.2f}, std: {std2:.2f}")

    # 3. Bar plot with error bars to visualize variability
    x = np.arange(2)
    means = [mean1, mean2]
    stds = [std1, std2]

    plt.figure(figsize=(6,4))
    plt.bar(x, means, yerr=stds, capsize=10, color=['skyblue','salmon'],
            alpha=0.7, edgecolor='k')
    plt.xticks(x, labels)
    plt.ylabel('Mean ± Std Dev')
    plt.title('Comparing Standard Deviations')
    plt.grid(axis='y', alpha=0.3)
    plt.show()

    return (mean1, std1), (mean2, std2)

if __name__ == "__main__":
    np.random.seed(0)
    d1 = np.random.normal(loc=50, scale=5, size=100)
    d2 = np.random.normal(loc=50, scale=10, size=100)
    compare_std(d1, d2, ('Low Variability', 'High Variability'))
```

24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

def analyze_skew_kurtosis(mu=0, sigma=1, n=1000):
    """
    Generates normal data and computes skewness and kurtosis.
    - skewness ≈ 0 for symmetric distribution
    - kurtosis ≈ 0 (excess kurtosis) for normal
    """
    data = np.random.normal(loc=mu, scale=sigma, size=n)
    s = skew(data, bias=False)
    k = kurtosis(data, fisher=True, bias=False)  # excess kurtosis
    print(f"Skewness: {s:.3f}")
    print(f"Excess kurtosis: {k:.3f}")

    # Visual check
    plt.figure(figsize=(6,4))
    plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='k')
    plt.title(f"Histogram: N({mu}, {sigma}²)\nSkewness={s:.3f}, Excess Kurtosis={k:.3f}")
    plt.xlabel('Value'), plt.ylabel('Density')
    plt.grid(alpha=0.3)
    plt.show()

    return s, k

if __name__ == "__main__":
    np.random.seed(42)
    analyze_skew_kurtosis(mu=0, sigma=1, n=5000)
```
